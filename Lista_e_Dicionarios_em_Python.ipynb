```python
#12
n = int(input("Quantas pessoas deseja cadastrar? "))

mais_velho = ""
idade_mais_velho = -1

for i in range(n):
    nome = input(f"Digite o nome da {i+1}ª pessoa: ")
    idade = int(input(f"Digite a idade de {nome}: "))

    if idade > idade_mais_velho:
        idade_mais_velho = idade
        mais_velho = nome

print(f"A pessoa mais velha é {mais_velho}, com {idade_mais_velho} anos.")
```

```python
#13
n = int(input("Quantas pessoas deseja cadastrar? "))

soma_m = soma_f = 0
cont_m = cont_f = 0

for i in range(n):
    sexo = input(f"Digite o sexo da {i+1}ª pessoa (M/F): ").upper()
    idade = int(input("Digite a idade: "))

    if sexo == "M":
        soma_m += idade
        cont_m += 1
    elif sexo == "F":
        soma_f += idade
        cont_f += 1

media_m = soma_m / cont_m if cont_m > 0 else 0
media_f = soma_f / cont_f if cont_f > 0 else 0

print(f"Média de idade dos homens: {media_m:.2f}")
print(f"Média de idade das mulheres: {media_f:.2f}")
```

```python
#16
 idades = []

while True:
    idade = int(input("Digite a idade do aluno: "))
    idades.append(idade)
    continuar = input("Deseja informar a idade do próximo aluno? (S/N): ").upper()
    if continuar != "S":
        break

mais_novo = min(idades)
mais_velho = max(idades)
maiores_18 = len([i for i in idades if i > 18])
menores_ou_18 = len([i for i in idades if i <= 18])
media = sum(idades) / len(idades)

idades.sort()
n = len(idades)
if n % 2 == 1:
    mediana = idades[n // 2]
else:
    mediana = (idades[n // 2 - 1] + idades[n // 2]) / 2

print(f"\nIdade do mais novo: {mais_novo}")
print(f"Idade do mais velho: {mais_velho}")
print(f"Alunos com mais de 18 anos: {maiores_18}")
print(f"Alunos com até 18 anos: {menores_ou_18}")
print(f"Média das idades: {media:.2f}")
print(f"Mediana das idades: {mediana}")
```

```python
alunos_db = [
    {"matricula": "2023001", "nome": "lucas", "curso": "Engenharia de softwere"},
    {"matricula": "2023002", "nome": "damaseno", "curso": "medicina"},
    {"matricula": "2023003", "nome": "silva", "curso": "direito"}
]
print("1. Sistema Simplificado de Cadastro de Alunos")
print ("_" * 40)
#1 .  Adicionar novo aluno
print ("_> Adicionar novo aluno...")
novo_aluno = {"matricula": "2023004", "nome": "ana", "curso": "Ciência da Computação"}
alunos_db.append(novo_aluno)
print(f"Aluno {novo_aluno['nome']} adicionado com sucesso")

matricula_busca = "2023002"
aluno_encontrado = next((aluno for aluno in alunos_db if aluno["matricula"] == matricula_busca), None)
if aluno_encontrado:
    print(f"\nAluno encontrado com matrícula: {matricula_busca}")
    #acessa os valores do dicionario usando suas chaves
    print(f"Nome: {aluno_encontrado['nome']}")
    print(f"Curso: {aluno_encontrado['curso']}")
else:
    print(f"\nAluno com matrícula {matricula_busca} não encontrado.")

#3. Atualizar informação do curso (update)
matricula_update = "2023003"
for aluno in alunos_db:
    if aluno["matricula"] == matricula_update:
        novo_curso = "Engenharia de Software"
        aluno["curso"] = novo_curso
        print(f"Curso do aluno {aluno['nome']} atualizado para {novo_curso}")
        break # Sai do loop após encontrar e atualizar o aluno
else:
    print(f"\nAluno com matrícula {matricula_update} não encontrado para atualização.")

print("\n--- Base de Alunos Atualizada ---")
for aluno in alunos_db:
    print(aluno)
```
